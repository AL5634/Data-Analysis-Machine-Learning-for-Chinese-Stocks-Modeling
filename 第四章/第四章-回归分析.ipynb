```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# 读取CSV数据
df = pd.read_csv('/Users/xiaoquanliu/Desktop/Book_DataCode/第四章/单变量回归案例.csv')
# 读取数据集并去除NaN值
data = df.dropna()
# 提取被解释变量和解释变量的数据
y = data["净资本自然对数"]
x = data["发布研报数量"]
# 添加常数列到解释变量数据中
x = sm.add_constant(x)
# 使用OLS函数拟合模型并进行回归分析
model = sm.OLS(y, x)
results = model.fit()
# 输出回归结果摘要
print(results.summary())

```

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

# 读取CSV数据
df = pd.read_csv('/Users/xiaoquanliu/Desktop/Book_DataCode/第四章/多元回归案例1.csv',encoding='gbk')

# Numpy数据格式转化
for col in ["Data", "Incentive_F", "Incentive_M","Senior_Edu","CEO_Edu","Ln_Age","TFP"]:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
# 数据清理
df_cleaned_any = df.dropna()

# 多元线性回归
X = df_cleaned_any[["Data", "Incentive_F", "Incentive_M","Senior_Edu","CEO_Edu","Ln_Age"]]
y = df_cleaned_any["TFP"]
X = sm.add_constant(X) 

model = sm.OLS(y, X)
results = model.fit()
print(results.summary())
```

```python


# 3. 分组回归（以是否为国有企业进行分组，显示分组多元回归结果变量）
for name, group in df_cleaned_any.groupby("SOE"):
    X = group[["Data", "Incentive_F", "Incentive_M","Senior_Edu","CEO_Edu","Ln_Age"]]
    y = group["TFP"]
    X = sm.add_constant(X)
    model = sm.OLS(y, X)
    results = model.fit()
    print(f"Regression results for group {name}")
    print(results.summary())


```

```python
from linearmodels import PanelOLS
import pandas as pd

#使用固定效用模型之前需要设定个体和时间的双重索引（two-level multiindex)
df.set_index(["cor", "year"], inplace=True)
# 指定regressors和dependent variable
exog_vars = ["vbankcr","cbankcr","vshortcr","vlongcr","clongcr","opcash","ppe","roa", "cshortcr", "size","lev","absnda","age"]
exog = sm.add_constant(df[exog_vars])
endog = df["rd"]
# 构建固定效应模型
model = PanelOLS(endog, exog, entity_effects=True)
# 进行回归
fe_res = model.fit()
print(fe_res)
```

```python

import pandas as pd
import statsmodels.api as sm

# 1. 本地导入数据
data = pd.read_csv("/Users/xiaoquanliu/Desktop/Book_DataCode/第四章/逻辑回归案例.csv")
# 2. 删除NaN值
data = data.dropna()  
# 分离因变量和自变量
X = data[["income","debt_r","FinanceCF","Inv","lAsset/Size","cash","Inv_income","OperationCF"]]
y = data["M&A"]
# 添加常数列作为回归模型的截距项
X = sm.add_constant(X)
# 构建逻辑回归模型
log_reg = sm.Logit(y, X)
result = log_reg.fit()
# 3. 输出结果
print(result.summary())

```
