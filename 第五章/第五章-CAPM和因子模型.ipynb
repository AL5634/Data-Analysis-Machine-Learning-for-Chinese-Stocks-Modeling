```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
# 读取CSV数据
df = pd.read_csv('/Users/xiaoquanliu/Desktop/Book_DataCode/第五章/CAPM案例数据.csv')
```

```python
df.head()
```

```python

# 提取股票和市场收益率数据
market_returns = df['恒生指数']

# 定义股票列表和字典存储结果
stocks = ['长和', '中移动', '汇丰', '腾讯']
results = {}

# 循环计算每支股票的beta和alpha
for stock in stocks:
    stock_returns = df[stock]
    market_returns = sm.add_constant(market_returns)
    
    model = sm.OLS(stock_returns, market_returns)
    result = model.fit()
    
    # 存储结果
    beta = result.params[1]
    alpha = result.params[0]
    results[stock] = {'beta': beta, 'alpha': alpha}

# 输出结果
for stock in stocks:
    print(f"{stock}的Beta系数: {results[stock]['beta']}")
    print(f"{stock}的Alpha: {results[stock]['alpha']}")
    print('---')

```
